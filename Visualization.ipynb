# Generating samples

```python
x = np.linspace(start=1, stop=8*np.pi, num=100)
sin_x = np.sin(x)
cos_x = np.cos(x)
exp_x = np.exp(x)
log_x = np.log(x)
```

# Figure and Subplots

plt.figure/plt.subplots

```python
plt.plot(x, sin_x, color='blue')
plt.plot(x, cos_x, color='green')
plt.show()
```

```python
plt.figure(figsize=(20,5))
plt.plot(x, log_x, color='blue', label="log")
plt.plot(x, sin_x, color='green', label="sine")
plt.legend(loc="upper left")
plt.title('Chart Title')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.xticks([x for x in np.linspace(start=1, stop=8*np.pi, num=20)])
plt.yticks([x for x in np.linspace(start=-1, stop=5, num=10)])
plt.show()
```

```python
fig, axs = plt.subplots(2,2)
fig.set_figheight(6)
fig.set_figwidth(15)
ax1, ax2, ax3, ax4 = axs[0,0], axs[0,1], axs[1,0], axs[1,1]
ax1.plot(x, sin_x)
ax1.set_title('Sine')
ax2.plot(x, cos_x, label="cos(x)")
ax2.plot(x, log_x, label="log(x)")
ax2.legend()
ax2.set_title('Cosine+Log')
ax3.plot(x, exp_x)
ax3.set_title('Exponential')
ax4.plot(x, log_x)
ax4.set_title('Logarithmic')
plt.subplots_adjust(hspace=0.3)
plt.show()
```

```python
fig = plt.figure(figsize=(15,6))
ax1 = fig.add_subplot(221) # plt.subplot2grid(shape=(2,2), loc=(0,0), rowspan=1, colspan=1)
ax2 = fig.add_subplot(222)
ax3 = fig.add_subplot(212)
ax1.plot(x, sin_x)
ax2.plot(x, cos_x)
ax3.plot(x, exp_x)
plt.subplots_adjust(wspace=0.1,hspace=0.2)
plt.show()
```

# Other types of charts
Scatter, Histogram, Bins

```python
np.random.seed(0)
m1 = np.random.random_sample(size=10)
m2 = np.random.random_sample(size=10)
```

```python
plt.scatter(m1,m2)
plt.xticks([x for x in np.linspace(start=0, stop=1, num=10)])
plt.yticks([x for x in np.linspace(start=0, stop=1, num=10)])
plt.show()
```

```python
seq = np.random.randint(low=1, high=6, size=10)
seq
```

```python
plt.hist(seq, bins=4, range=(1,5), align='mid', rwidth=0.8)
plt.show()
```

```python
bin_edges=[1,3,5]
counts, bins, patches = plt.hist(seq, bins=bin_edges, rwidth=1)
plt.xticks(bins)
plt.show()
```
